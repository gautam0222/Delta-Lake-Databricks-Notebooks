### **SCHEMA EVOLUTION**

```python
data = [
    (1,100,"aa"),
    (2,200,"bb"),
    (3,300,"cc")
]

df = spark.createDataFrame(
    data,
    ["id","salary","name"]
)

path="/Volumes/deltalakegautam/default/deltavol/demosink"

df.write \
.format("delta") \
.mode("overwrite") \
.save(path)
```

```python
new_data = [
    (4,400,"dd","Nagpur"),
    (5,500,"ee","Mumbai")
]

new_df = spark.createDataFrame(
    new_data,
    ["id","salary","name","city"]
)

display(new_df)
```

```python
new_df.write \
.format("delta") \
.mode("append") \
.save(path)
```

```python
new_df.write \
.format("delta") \
.mode("append") \
.option("mergeSchema","true") \
.save(path)
```

```python
result = spark.read \
.format("delta") \
.load(path)

display(result)
```

# Schema overwrite

```python
data = [
    (1,100,"aa",1),
    (2,200,"bb",1),
    (3,300,"cc",1),
    (4,400,"dd",1)
]

df = spark.createDataFrame(
    data,
    ["id","salary","name","tip"]
)

display(df)
```

```python
df.write.format("delta")\
        .mode("overwrite")\
        .option("overwriteSchema","true")\
        .save("/Volumes/deltalakegautam/default/deltavol/demosink/")
```
