## Using SQL TABLE

```python
%sql
create table deltalakegautam.default.firsttable(
    id int,
    salary int
);
```

### **Using Delta API**

```python
from delta.tables import DeltaTable, IdentityGenerator
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
DeltaTable.createIfNotExists(spark) \
  .tableName("deltalakegautam.default.firstdeltaapi") \
  .addColumn("id", "INT") \
  .addColumn("salary", "INT") \
  .execute()
```

```python
%sql
drop table deltalakegautam.default.firstdeltaapi
```

### **Identity Columns**

```python
DeltaTable.create(spark)\
    .tableName("deltalakegautam.default.firstdeltaapi")\
    .addColumn("id",dataType = LongType(), generatedAlwaysAs= IdentityGenerator())\
    .addColumn("salary", dataType = IntegerType())\
    .addColumn("name", dataType = StringType())\
    .execute()
```

### **Computed Columns**

```python
%sql
insert into deltalakegautam.default.firstdeltaapi(salary, name) values
(1000, 'Gautam'),
(2000, 'Ravi'),
(3000, 'Raj'),
(4000, 'Rajesh');
```
