# DML

```python
data = [
    (1,100,"aa",1),
    (2,200,"bb",1),
    (3,300,"cc",1),
    (4,400,"dd",1)
]

df=spark.createDataFrame(data,["id","salary","name","tip"])

df.write.format("delta")\
        .mode("append")\
        .save("/Volumes/deltalakegautam/default/deltavol/dmlsink/")
```

```python
%sql
select * from delta.`/Volumes/deltalakegautam/default/deltavol/dmlsink/`
```

```python
%sql
update delta.`/Volumes/deltalakegautam/default/deltavol/dmlsink/` set salary = 1000000 where id = 5
```

# Upsert Command

```python
data = [
    (1,100,"aa",1),
    (9,200,"bb",1),
    (10,300,"cc",1),
]

df=spark.createDataFrame(data,["id","salary","name","tip"])

display(df)
```

```python
from delta.tables import DeltaTable
```

```python
dlt_obj = DeltaTable.forPath(spark, "/Volumes/deltalakegautam/default/deltavol/dmlsink/")
dlt_obj.alias("trg").merge(
    df.alias("src"),
    "trg.id = src.id"
).whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
```
